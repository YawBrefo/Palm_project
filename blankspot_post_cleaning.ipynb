```python
# Post blankspot cleaning script

# Import required libraries
import geopandas as gpd

# function to edit palm count after QA
def blankspot_edit(blank_dir, blk_number, dest_dir):
    
    # read count geojson file
    blank_gdf = gpd.read_file(blank_dir)

    # remove any duplicate point
    blank_gdf = blank_gdf.drop_duplicates(subset='geometry')

    # set plant id
    blank_gdf['Plant_id'] = (blank_gdf.index + 1).astype(str)

    # Replace coords with geometry where it's null
    blank_gdf['lon'] = blank_gdf['lon'].fillna(blank_gdf['geometry'].x)
    blank_gdf['lat'] = blank_gdf['lat'].fillna(blank_gdf['geometry'].y)

    # set count geometry
    blank_gdf['geometry'] = gpd.points_from_xy(blank_gdf['lon'], blank_gdf['lat'])

    # assign block number
    blank_gdf['block'] = blk_number

    # save the geodataframe as geojson file
    blank_gdf.to_file(f'{dest_dir}final_blankspot_points.geojson', driver='GeoJSON')


#read blankspot file
blank_dir = '.../Project_files/post_QA_blankspots.geojson'
blk_number = '1' # set block number for the blocks
dest_dir = '.../Project_repo/blankspot/blankspot_post_QA/'


blankspot_edit(blank_dir, blk_number, dest_dir)
```
