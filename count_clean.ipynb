```python
# libraries
import geopandas as gpd
from pathlib import Path
```

After the point geolocation files have been generated and manually visualized for accuracy,
a clean-up script works on to generate Palm_id, block id, lat, long and save as geojson.

```python
def post_QA_edit(in_count_geojson, out_count_geojson, block_id):
  # read in count files
  for count_file in Path(in_count_geojson).iterdir():
    gdf = gpd.read_file(count_file)

    # remove any duplicates
    gdf = gdf.drop_duplicates(subset='geometry')

    # rename and number the plant_id for count
    gdf['Plant_id'] = (gdf.index + 1).astype(str)
    gdf['block_id'] = block_id

    # Replace 'lat' with geometry y where 'lat' is null
    gdf['lat'] = gdf.geometry.y

    # Replace 'lon' with geometry x where 'lon' is null
    gdf['long'] = gdf.geometry.x

    # compute diameter for non-existence
    gdf['diameter'] = gdf['diameter'].fillna(gdf['diameter'].mean())
    gdf['diameter'] = round(gdf['diameter'], 2)

    # save the geodataframe as geojson file
    gdf.to_file(f'{out_count_geojson}{count_file.stem}.geojson', driver='GeoJSON')

    print(f'{out_count_geojson}{count_file.stem}.geojson, has been post_processed')

# Paths
in_count_dir = '.../Project_repo/count/count_geo_output/' # point_files after QA
out_count_dir = '.../Project_repo/count/count_post_QA_output/' # final count_geojson
block_num = 2 # block number/id

# Implementation
post_QA_edit(in_count_dir, out_count_dir, block_num)
```

```python
# All block QA'd count files
def value_(in_count_path, in_block_count_path):

    # Read QA'd count file
    count_gdf = gpd.read_file(in_count_path)
    
    # filter blocks form merged count
    for block_id, gdf_block in count_gdf.groupby('block'):
        print(f"Block: {block_id}, Trees: {len(gdf_block)}")

        # save the geodataframe as geojson file
        gdf_block.to_file(f'{in_block_count_path}{block_id}.geojson', driver='GeoJSON')


blankspot_dir = '/..../blank_spots.geojson'
boundary_dir = '/..../boundary_files/'
block_blankspot_dir = '/..../separate_QA_blankspot/'

value_(blankspot_dir, block_blankspot_dir)
```
